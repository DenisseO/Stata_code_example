## Global Poverty Measures

* The Gini Coefficient is the most commonly used measure ofinequality. It captures the deviation from the line of equality.
* The Lorenz Curve shows the income distribution.

```stata
clear all                         
set more off, permanently         
set matsize 800
```

Install the following packages: glcurve, ineqdeco, ineqdec0.

```stata
ssc install glcurve
ssc install ineqdeco 
ssc install ineqdec0
```

### Data

```stata
use ELP.dta
```

### Lorenz Curves

```stata
tab country
```

```stata
glcurve consumption_tot if country=="Gotham" [aw=weight_final], glvar(g1) pvar(p1) nograph lorenz replace
```

```stata
foreach country in "Gotham" "Metropolis" "Sterling" "Wakanda" "Zamunda"{
    qui glcurve consumption_tot if country=="`country'" [aw=weight_final], glvar(g`country') pvar(p`country') nograph lorenz replace 
    twoway (line g`country' p`country', sort) (function y=x ,range(0 1)), graphregion(fcolor(white))title("Lorenz Curve `country'")ytitle("Income (Cumulative Percent)")legend(label(1 "`country'") label(2 "Perfect equality"))
    graph export "graphs/`country'.jpg"
}
```

### GINI Coefficient 

Integration method 

```stata
foreach country in "Gotham" "Metropolis" "Sterling" "Wakanda" "Zamunda"{
    qui integ g`country' p`country'
    display "`country' =" (1-2*r(integral))
}
```

Using ineqdec0 ineqdeco

```stata
ineqdeco consumption_total1 [aw=weight_final1], bygroup(fe)
```
