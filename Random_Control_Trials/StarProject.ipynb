# STAR PROJECT

The aim of this post is to give Econometrics student´s an explanation and code example in Stata of a social experiment in economics using RANDOMIZED CONTROLLED TRIALS. The following work is base on the results of the analysis on [KRUEGER](https://www.jstor.org/stable/2587015?seq=1) paper´s. We will use the START dataset. It contains information about 11,600 students and their teachers. The students where randomly assigned to different size classes from kindergarten through third grade (Krueger,1999)

I used Jupyter Notebooks to show how this tool can be a good way to present and explain results in Stata. 

### Description
We going to analysis and generate the table 1 describe in [KRUEGER](https://www.jstor.org/stable/2587015?seq=1) paper´s. The amin of the following explanation is to prove that the characteristics of treatment and control groups are similar.

### Data

We will use the following variables from the STAR project data :

* gkclasstype: Classroom type kindergarten
  * small class (13–17)
  * regular class (22–25 students)
  * regular plus a full-time aide class
* gkfreelunch: Free/reduced lunch status in kindergarten
* race: Student race/Ethnicity
* birthyear: Student year of birth


```stata
use STAR_Students, clear
```

### Gen. dummies variables

```stata
tab gkfreelunch, gen(gkflunch)
```

```stata
tab gkclasstype, gen(gktype)
```

```stata
tab race // white and Asian 
gen whiteasian = 0 
replace whiteasian=1 if race==1
replace whiteasian=1 if race==3
```

### Run separate regressions
Baseline variables on the Treatment indicator only

```stata
reg gkflunch1 gktype1 gktype2 gktype3, r noconstant
```

### F-test of equality of all three groups

```stata
test gktype1 = gktype2 = gktype3
```

### Repeat this process for the others baseline variables

```stata
global balance gkflunch1 whiteasian gkclasssize
global gk gktype1 gktype2 gktype3
```

```stata
foreach y in $balance{
    eststo: qui reg `y' $gk, r noconstant
    estimates store `y'
}
```
